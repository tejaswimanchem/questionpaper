```python
from matplotlib import pyplot as plt
x = [2,4,6,8,9,12,14]
y = [1,4,2,5,3,7,8]
plt.plot(x,y)
plt.show()
```

```python
from matplotlib import pyplot as plt
x = [2,4,6,8,9,12,14]
y = [1,4,2,5,3,7,8]
plt.bar(x,y)
plt.show()
```
